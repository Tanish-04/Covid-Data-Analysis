# Data Extraction through Repositories

### Global Recovered Cases

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
recovered_global_cases = pd.read_csv(url)
recovered_global_cases.head(5)
```

### Global Deaths Cases

```python
url1 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
death_global_cases = pd.read_csv(url1)
death_global_cases.head(5)
```

### Global Confirmed Cases

```python
url2 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
confirmed_global_cases = pd.read_csv(url2)
confirmed_global_cases.head(5)
```

   # Data Preparation

### Recovered Global Cases

```python
recovered_global_cases.drop(['Lat','Province/State','Long'], axis=1, inplace=True)
recovered_global_cases.head()
```

```python
groupingOfRecoveredCases = recovered_global_cases.pivot_table(index = ["Country/Region"])
groupingOfRecoveredCases.head()
```

```python
Recovered_Global_Cases = groupingOfRecoveredCases.transpose()
Recovered_Global_Cases.head()
```

### Death Global Cases

```python
death_global_cases.drop(['Lat','Province/State','Long'], axis=1, inplace=True)
death_global_cases.head()
```

```python
groupingOfDeathCases = death_global_cases.pivot_table(index = ["Country/Region"])
groupingOfDeathCases.head()
```

```python
Death_Global_Cases = groupingOfDeathCases.transpose()
Death_Global_Cases.head()
```

### Confirmed Global Cases

```python
confirmed_global_cases.drop(['Lat','Province/State','Long'], axis=1, inplace=True)
confirmed_global_cases.head()
```

```python
groupingOfGlobalCases = confirmed_global_cases.pivot_table(index = ["Country/Region"])
groupingOfGlobalCases.head()
```

```python
Confirmed_Global_Cases = groupingOfGlobalCases.transpose()
Confirmed_Global_Cases.head()
```

##Task 04
Now we need to merge the dataframes. One way to achieve this is suggested below, but you are free to explore your own methods.
1. Add a column named "Status" to your each dataframe containing **confirmed, death,** and **recovered** for confirmed_global_cases, death _global_case, recovered_global_cases respectively.
2. Converet your Date column to DateTimeIndex
3. Use a suitable merge to join the 3 data frames.

```python
Confirmed_Global_Cases['Status'] = "confirmed"
Death_Global_Cases['Status'] = "death"
Recovered_Global_Cases['Status'] = "recovered"
Confirmed_Global_Cases
```

```python
Confirmed_Global_Cases.reset_index(inplace=True)
Death_Global_Cases.reset_index(inplace=True)
Recovered_Global_Cases.reset_index(inplace=True)
```

```python
Confirmed_Global_Cases['index'] = pd.to_datetime(Confirmed_Global_Cases['index'])
Death_Global_Cases['index'] = pd.to_datetime(Death_Global_Cases['index'])
Recovered_Global_Cases['index'] = pd.to_datetime(Recovered_Global_Cases['index'])
```

```python
df = Confirmed_Global_Cases.sort_values(['index']).set_index(['index'])
df1 = Death_Global_Cases.sort_values(['index']).set_index(['index'])
df2 = Recovered_Global_Cases.sort_values(['index']).set_index(['index'])
```

```python
merging = df.append(df1)
merging_data = merging.append(df2)
```

```python
final_result = merging_data.groupby(["index","Status"]).agg(np.average)
final_result
```

# Visualization

Top 10 Countries List where
1) Most Confirmed Cases of Covid
2) Most Deaths
3) Most Recovered Cases

```python
col_list = list(merging_data)

Countrywise_sum_of_confirmed_cases = merging_data[merging_data["Status"] == "confirmed"][col_list].sum().drop('Status')
top10_countries_confirmedcasessum = Countrywise_sum_of_confirmed_cases.sort_values(ascending=False).head(10)

Countrywise_sum_of_deaths = merging_data[merging_data["Status"] == "death"][col_list].sum().drop('Status')
top10_countries_deathsum =Countrywise_sum_of_deaths.sort_values(ascending=False).head(10)

Countrywise_sum_of_recovered_cases = merging_data[merging_data["Status"] == "recovered"][col_list].sum().drop('Status')
top10_countries_recoveredcasessum = Countrywise_sum_of_recovered_cases.sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 5))
top10_countries_confirmedcasessum.plot(kind='bar', grid=True)
top10_countries_recoveredcasessum.plot(color='r',  grid=True, marker='o')
top10_countries_deathsum.plot(color='g', grid=True, marker='o')
plt.legend(['Recovered Cases','Death Cases','Confirmed Cases'])
```
